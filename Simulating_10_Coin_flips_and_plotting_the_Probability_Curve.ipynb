```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
#Number of coin flips per experiment

n = 10

#Number of times we repeat the experiment

number_of_experiments = 10000

#Simulating the experiment

result = []

for i in range(number_of_experiments):

  no_of_flips = np.random.choice([0, 1], size = n, p = [0.5, 0.5])  #0 for heads, 1 for tails

  #Counting the number of heads

  number_of_heads = np.sum(no_of_flips)

  result.append(number_of_heads)


print("Simulated Result")
print(result)

#Counting occurences of each possible numebr of heads(0 to 10)

values, counts = np.unique(result, return_counts = True)

#Converting counts to Probabilities

probability = counts / number_of_experiments
```

```python
#Plotting

plt.figure(figsize =(8,5))

plt.bar(values, probability, color = 'blue', edgecolor = 'black', alpha = 0.7, label = "Simulated Probability")

plt.plot(values, probability, 'o-', color = 'lightcoral', label = "Smooth Curve")  #Connecting dots for Smoothness.

#Labels and title

plt.xlabel("Number of heads in 10 coin flips")

plt.ylabel("Probability")

plt.title("Simulated Probability Distribution of Number of Heads in 10 Coin Flips")

plt.xticks(np.arange(0, n + 1, 1))

plt.legend()

plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

plt.show()
```
